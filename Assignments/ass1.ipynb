```python
# Problem 1

# Apply all the preprocessing techniques that you think are necessary

import re 
def html_tag_remover(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'',text)
```

```python
df['review'] = df['review'].apply(html_tag_remover)
```

```python
df['review'] = df['review'].str.lower()
```

```python
df['review']
```

```python
def remove_punchuation(text):
    for char in exclude:
        text = text.replace(char,'')
    return text
```

```python
import string, time 
exclude = string.punctuation
```

```python
df['review']=df['review'].apply(remove_punchuation)
```

```python
# Problem 4

# Apply bag words and find the vocabulary also find the times each word has occured
```

```python
cv = CountVectorizer(max_features=500) 
BoW=cv.fit_transform(df['review'])
```

```python
print(cv.vocabulary_)
```

```python
#  find the times each word has occured
import numpy as np 
words = cv.get_feature_names_out() 

counts = np.array(BoW.sum(axis=0)).flatten() 

for word, count in zip(words, counts):
    print(word, ':', count)
```

```python
# Problem 6

# Apply tf-idf and find out the idf scores of words, also find out the vocabulary.
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer 

tfidf=TfidfVectorizer(max_features=500)

x= tfidf.fit_transform(df['review']) 

print(tfidf.vocabulary_)
```

```python
print(tfidf.idf_ ) 
print(tfidf.get_feature_names_out())
```

```python
idf_score = tfidf.idf_ 
words = tfidf.get_feature_names_out() 

for word, score in zip(words, idf_score):
    print(word, ":", score)
```
