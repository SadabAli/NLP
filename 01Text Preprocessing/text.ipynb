## Removing URL

```python
def removing_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)
```

## Removing Punchuation

```python
import string,time
```

```python
exclude = string.punctuation
exclude
```

### From scratch

```python
def remove_punchuation(text):
    for char in exclude:
        text = text.replace(char,'')
    return text
```

```python
msg='I LOVE YOU soooo much!!!!!!!!!'
```

```python
remove_punchuation(msg)
```

```python
start = time.time()
print(remove_punchuation(msg))
time1 = time.time() - start
print(time1*50000)
```

### Modern way

```python
def remove_punc1(text):
    return text.translate(str.maketrans('', '', exclude))
```

```python
text='I am so Youholick!!!:('
remove_punc1(text)
```

```python
start = time.time()
remove_punc1(msg)
time2 = time.time() - start
print(time2*50000)
```

## Removing Stopwords

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('English')
```

```python
def remove_stopwords(text):
    new_text = []
    
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
remove_stopwords('probably my all-time favorite movie, a story of selflessness, sacrifice and dedication to a noble cause, but it\'s not preachy or boring. it just never gets old, despite my having seen it some 15 or more times')
```

## Handling Imojies

```python
import re
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
remove_emoji("Loved the Rohit Sharma😘😘")
```
